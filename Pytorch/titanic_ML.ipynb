# Modeling and Learning  
To learn, convert to data into array form


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
x_train = data_train.to_numpy()
y_train = survived_train.to_numpy()

x_val = data_val.to_numpy()
y_val = survived_val.to_numpy()
```

numpy array to tensor

```python
x_train = torch.FloatTensor(x_train)
y_train = torch.FloatTensor(y_train)
x_val = torch.FloatTensor(x_val)
y_val = torch.FloatTensor(y_val)
```

```python
a = torch.tensor([0,2])
b = torch.tensor([1,1])
torch.cat((a,b), dim = 0)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(8,64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64,32)
        self.fc3= nn.Linear(32,1)
        self.sigmoid = nn.Sigmoid()
    def forward(self,x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        
        return x
```

# Tuning hyperparameters using validation

```python
net = Net()
optimizer = optim.Adam(lr=0.001,params= net.parameters())
```

```python
x_train
```

# Accuracy and test

I think it is too small to get good enough result due to tiny dataset.  
So I decide not to split train and valid set

```python
x_set = torch.cat((x_train, x_val),dim = 0)
y_set = torch.cat((y_train, y_val),dim = 0)
```

```python
net = Net()
optimizer = optim.Adam(lr=0.0005,params= net.parameters())
```

```python
epochs = 200

for epoch in range(epochs):
    running_loss = 0.0
    running_val_loss = 0.0
    accuracy = 0.0
    running_loss = 0.0
    for i, (x,y) in enumerate(zip(x_set,y_set)):
        
        y = torch.tensor([y])
        optimizer.zero_grad()
        outputs = net(x)

        loss = F.binary_cross_entropy(outputs,y)
        loss.backward()
        optimizer.step()
        running_loss+=loss.item()
        y_hat = outputs[0]
        if y_hat>0.5:
            y_hat = 1
        else:
            y_hat = 0
        #print(y_train.shape)
        if y_hat == y[0]:
            accuracy+=1
    print("[{:2d},{:2d}] accuracy : {:.3f}, loss : {:.3f}".format(epoch,epochs,accuracy/len(x_set),running_loss))
    
```
